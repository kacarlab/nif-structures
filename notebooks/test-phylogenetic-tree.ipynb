```python
def plot_tree(tre, reference_tips, reference_nodes, query_tip, query_node):

    rtre = tre .root(wildcard='BchChl')
    rtre = rtre.drop_tips(wildcard="BchChl")
    # rtre = rtre.drop_tips(wildcard="Nif")
    tip_labels = []
    tip_colors = []
    node_sizes = []
    node_labels = []

    for node in rtre.get_node_values('name', False, False):
        if node == query_node:
            print("conga!")
            node_labels.append(node)
        elif node in reference_nodes:
            print("bingo!!")
            node_labels.append(node)
        else:
            node_labels.append('')
    
    print(node_labels)

    for node in rtre.get_tip_labels():
        if node == query_tip:
            node_sizes.append(10)
        elif node in reference_tips:
            node_sizes.append(10)
        else:
            node_sizes.append(0)


    for node in rtre.get_tip_labels():
        if node == query_tip:
            tip_labels.append(node)
        elif node in reference_tips:
            tip_labels.append(node)
        else:
            tip_labels.append('')

    for node in rtre.get_tip_labels():

        if node == query_tip:
            tip_colors.append('red')
        elif node in reference_tips:
            tip_colors.append('white')
        else:
            tip_colors.append('')

    return rtre.draw(
        tip_labels_align=False, 
        tree_style='d', 
        tip_labels=tip_labels, 
        tip_labels_colors=tip_colors, 
        node_labels=node_labels,
        node_sizes=40
    )
```

```python
tre = toytree.tree(open('../data/tree/AGNifAlign103.asr.tre').read(), tree_format=1)

nodes = [node for node in tre.treenode.traverse()]
for node in nodes: 
    if node.is_leaf() is False:
        node.name = 'Anc_' + node.name

tre = tre.root(wildcard="BchChl_")
# tre = tre.drop_tips(wildcard="BchChl_")
nodes = [node for node in tre.treenode.traverse("levelorder")]

relationships = []
for node in nodes:
    for anc_node in node.get_ancestors():
        relationships.append(
            dict(type='ancestor', x=node.name, y=anc_node.name)
        )
    for anc_node in node.get_descendants():
        relationships.append(
            dict(type='succesor', x=node.name, y=anc_node.name)
        )
    
```

```python
an = pd.DataFrame.from_records(relationships).query('x == "Anc_802"').query('type == "succesor"')['y'].to_list()
nif3 = pd.DataFrame.from_records(relationships).query('x == "Anc_820"').query('type == "succesor"')['y'].to_list()
nif1 = pd.DataFrame.from_records(relationships).query('x == "Anc_1207"').query('type == "succesor"')['y'].to_list()
nif2 = pd.DataFrame.from_records(relationships).query('x == "Anc_1345"').query('type == "succesor"')['y'].to_list()


nif3 = [x for x in nif3 if x not in an]
print(len(nif1))
print(len(nif2))
print(len(nif3))
print(len(an))
print(len(an) + len(nif3) + len(nif2) + len(nif1))


clade_asignments = []

for item in an:
    clade_asignments.append(dict(name=item, clade='an'))
for item in nif3:
    clade_asignments.append(dict(name=item, clade='nif-iii'))
for item in nif2:
    clade_asignments.append(dict(name=item, clade='nif-ii'))
for item in nif1:
    clade_asignments.append(dict(name=item, clade='nif-i'))

clade_asignments = pd.DataFrame.from_records(clade_asignments)
clade_asignments.set_index('name').to_dict()
```
