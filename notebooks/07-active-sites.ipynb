```python
import pandas as pd
import seaborn as sns
```

```python
rmsd_femoco = pd.read_json('../structures/rmsd-femoco.json')
rmsd_fe8s7 = pd.read_json('../structures/rmsd-fe8s7.json')
```

```python
rmsd_fe8s7['cofactor'] = 'Fe8S7'
rmsd_femoco['cofactor'] = 'FeMo-co'
```

```python
rmsd = pd.concat([rmsd_fe8s7, rmsd_femoco])
```

```python
g = sns.displot(data=rmsd, hue='cofactor', x='value', bins=20, height=3.0)
g.set_xlabels('RMSD ($\AA$)')
g.savefig('../figures/ddkk-hh-histogram-active-site-RMSD.svg')
g.savefig('../figures/ddkk-hh-histogram-active-site-RMSD.png', dpi=300)
```
